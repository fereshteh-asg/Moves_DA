```python
import trj_builder
import scipy.stats as stats
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from pylab import plot,show,hist
import json
import datetime, time
from scipy.stats.kde import gaussian_kde
from scipy.stats import norm
from numpy import linspace,hstack
import numpy
import scipy
fmt="%Y%m%dT%H%M%S"
%matplotlib inline
```

```python
json_file = "raw_data.json"
with open(json_file, "r")  as jsonf:
    data_dict = json.load(jsonf)
###============================================

t_list=[]
for user,v in data_dict.iteritems():
    user_trj=0
    #print v["info"]["name"]
    for day in v["data"]:
        if (day!={}):
            dist_list,total_trj,diff=trj_builder.trajectory_builder(v["data"][day])
            for i in range(0,len(total_trj)):
                s_t=total_trj[i][0]['time'] 
                e_t=total_trj[i][-1]['time'] 
                s_t= datetime.datetime.strptime(s_t.split('+')[0],fmt)
                e_t= datetime.datetime.strptime(e_t.split('+')[0],fmt)  
                diff= e_t-s_t
                t_list.append(diff.seconds/60)
                user_trj+=1
print 'total number of trajectories:',len(t_list)
fig = plt.figure(figsize=(8,6))
plt.hist(t_list, bins=300, normed=True)
plt.xlim(0,400)
plt.xlabel('Trajectory Duration (Minutes)')
plt.show()
```

```python
fig = plt.figure(figsize=(8,6))
plt.hist(t_list,normed=True,bins=500,alpha=.4, color='g',label= '')
plt.xlim(0,160)
pdf = gaussian_kde(t_list)
cdf = stats.binom.cdf
x = linspace(0,160,300)
plot(x,pdf(x), 'b')
plt.xlabel("Trajectory Duration (Minutes)")
plt.title("Time distribution of trajectories")
plt.ylabel("Probability")
plt.legend()
plt.savefig("time_dist",format='pdf')
plt.show()
```
